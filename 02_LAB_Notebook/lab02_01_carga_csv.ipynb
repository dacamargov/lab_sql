### Parametros Iniciales

```python
%python
import pandas as pd
from pyspark.sql import SparkSession

url = 'https://raw.githubusercontent.com/Gabriel-Rangel/lab_sql/refs/heads/main/dados/'
catalog_name = f"dbacademy"
```

```python
%python

entity_name  = f"stock_bigtech"

table_name   = f"{catalog_name}.{schema_name}.{entity_name}"
file_name = f"{url}{entity_name}.csv"

df = pd.read_csv(file_name)                             # lectura de archivo CSV utilizando Dataframe Pandas
s_df = spark.createDataFrame(df)                        # convierte Dataframe Pandas en Spark Dataframe
s_df.write.mode("overwrite").saveAsTable(table_name)    # graba el DataFrame en la Tabla Delta
```
